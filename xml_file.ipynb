# how to create a xml file with the csv

```python
import csv
import xml.etree.ElementTree as ET

# Function to convert CSV data to XML
def csv_to_xml(csv_file, xml_file):
    # Read CSV data
    with open(csv_file, 'r') as csvfile:
        csv_reader = csv.DictReader(csvfile)
        data = list(csv_reader)

    # Create XML structure
    root = ET.Element('SDPSyncMessage')
    root.set('xmlns', 'http://www.emeter.com/energyip/syncinterface/v8')
    root.set('xmlns:xsi', 'http://www.w3.org/2001/XMLSchema-instance')

    header = ET.SubElement(root, 'header')
    for key, value in data[0].items():
        ET.SubElement(header, key).text = value

    payload = ET.SubElement(root, 'payload')
    device = ET.SubElement(payload, 'device')

    for row in data[1:]:
        parameter = ET.SubElement(device, 'parameter')
        for key, value in row.items():
            ET.SubElement(parameter, key).text = value

    # Create ElementTree object
    tree = ET.ElementTree(root)

    # Write to XML file
    with open(xml_file, 'wb') as xmlfile:
        tree.write(xmlfile)

# Example usage: csv_to_xml('your_input.csv', 'output.xml')
```
